### Packages

```python
import numpy as np

import pandas as pd

from shapely.geometry import Point

import matplotlib.pyplot as plt

import random
```

### Environment

### Model 1


#### Agent

UAV

#### State

For uav $i$, its observation is $o_{i}=(x_{i},y_{i})$


The overall state is $O=\{o_{1},o_{2}\}$


#### Action

For uav $i$,  its action is $(v_{i},\theta_{i})$


### Model 2

#### Agent

the control system

#### State

The state is  $S=\{ o_{1},o_{2}\}$ with information of all users


#### Action

For uav $i$,  its action is $(v_{i},\theta_{i})$

```python
class UE:
    
    def __init__(self,x,y,T=60, agent_num=2):
        
        self.D = np.zeros(T)  
        self.F = np.zeros(T)  
        self.L = np.zeros((agent_num, T))  
        self.fl = 1e9  
        self.x = x 
        self.y = y
        

class UAV:
    
    def __init__(self,x,y, T=60):
        
        self.D = np.zeros(T)  
        self.Theta = np.zeros(T)
        self.x = x
        self.y = y
        self.X = np.zeros(T) 
        self.Y = np.zeros(T) 
        self.X[0] = x
        self.Y[0] = y
        self.fc = 12e9  

class environment(object):
    
    def __init__(self):
        
        self.x_max = 400
        
        self.y_max = 400
        
        self.d_max = 30 
        
        self.H = 75  
        
        self.R_max = 75
        
        self.agent_num = 2 # the number of agents
        
        self.T = 60
        
        self.activate_user()
        
    def activate_user(self):
        
        # Functions: activate the users in the system
        
        # Read the user data
        
        user_df=pd.read_csv('users.csv',header=None,names=['x','y'])
        
        self.users=list()
        
        for idx,row in user_df.iterrows():
            
            self.users.append(UE(row.x,row.y))
            
    def reset(self):
        
        # Functions: reset the spatial coordinates of agent
        
        self.uavs=list()
        
        for a in range(self.agent_num):
            
            x,y=200,200
            
            self.uavs.append(UAV(x,y))
    
    
    def get_state(self,t):
        
        # Functions: get the current state
        
        state=list()
        
        for a in range(self.agent_num):
            
            x,y=self.uavs[a].x,self.uavs[a].y
            
            state.append([x,y])
            
        return state
            

    def ramdom_action(self):
        
        # Functions: randomly sample the action for an agent
        
        action=list()
        
        for a in range(self.agent_num):
            
            theta,d=np.random.random()*2*np.pi,np.random.random()*self.d_max
            
            action.append([theta,d])
            
        return action
    
    def transmit(self,distance):
        
        Bw = 5e6
        
        noise = 1e-12
        
        G_0 = 2.2846
        
        g_0 = 1.42e-4
        
        alpha = (G_0*g_0)/noise 
        
        return Bw*np.log2(1+(alpha*0.1)/(pow(self.H,2)+pow(distance,2)))
        
    
    def step(self,action,t):
        
        # Functions: update the next state and get the reward
        
        # 00 update user preference
        
        for u in range(len(self.users)):
            
            self.users[u].D[t]=np.random.choice([5e6, 10e6])
            
            self.users[u].F[t]=np.random.choice([1000e6, 1200e6])
            
            random_preference=np.random.random(self.agent_num)
            
            sum_preference=random_preference.sum()
            
            for a in range(self.agent_num):
            
                self.users[u].L[a][t]=random_preference[a]/sum_preference
        
        # 01 Service distance

        distance=np.zeros([self.agent_num,len(self.users)])
        
        for a in range(self.agent_num):
            
            for u in range(len(self.users)):
                
                distance[a][u]=Point(self.uavs[a].x,self.uavs[a].y).distance(Point(self.users[u].x,self.users[u].y))
                
        covers=np.where(distance<=self.R_max,1,0)
        
        # 02 transmission rate
        
        transmission=np.zeros([self.agent_num,len(self.users)])
        
        for a in range(self.agent_num):
            
            for u in range(len(self.users)):
                
                transmission[a][u]=self.transmit(distance[a][u])
                
        # 03 calculate the reward
        
        reward=list()
        
        for a in range(self.agent_num):
            
            uav_cost=0
            
            served_users_num=covers[a].sum()
            
            fc_a = self.uavs[a].fc /served_users_num
            
            for u in range(len(self.users)):
                
                l_au =self.users[u].L[a][t]
                
                D_u = self.users[u].D[t]
                
                F_u = self.users[u].F[t]
                
                rate = transmission[a][u]
                
                fl_u  = self.users[u].fl
                
                EC_au= 0.1 * D_u/rate
                
                EL_u = 1e-28 * pow(fl_u, 2) * F_u
                
                uc_u = np.linalg.norm(covers[:,u], ord=1)
                
                if uc_u == self.agent_num:
                    
                    uc_u = 1
                    
                else:

                    uc_u = 0
                    
                cost = l_au * ((1 - covers[a][u]) * EC_au + uc_u * EL_u)
                
                uav_cost+=cost
                
            reward.append(-1*uav_cost)
            
        # 04 Update the location of UAV
        
        for a in range(self.agent_num):
            
            agent_action=action[a]
            
            theta,d=agent_action[0],agent_action[1]
            
            x,y=self.uavs[a].x,self.uavs[a].y
            
            x_,y_=x+d*np.cos(theta),y+d*np.sin(theta)
            
            x_,y_=np.clip(x_, 0, self.x_max),np.clip(y_, 0, self.y_max)
            
            # store
            
            self.uavs[a].D[t]=d
            
            self.uavs[a].Theta[t]=theta
            
            self.uavs[a].x=x_
            
            self.uavs[a].y=y_
            
            if t+1<=self.T-1:
            
                self.uavs[a].X[t+1]=x_
            
                self.uavs[a].Y[t+1]=y_
            
            
        # 05 done
        
        done=False
        
        if t>=self.T-1:
            
            done=True
            
        return reward,done
        
             
```

### Simulation

```python
env=environment()

done=False

t=0

env.reset()

episode_reward=list()

while not done:
    
    action=env.ramdom_action()
    
    reward,done=env.step(action,t)
    
    episode_reward.append(np.array(reward).mean())
    
    t+=1
    
np.array(episode_reward).mean()
```
